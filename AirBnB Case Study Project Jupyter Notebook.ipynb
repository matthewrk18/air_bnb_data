```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

nG = air_bnb.neighbourhood_group.value_counts()
print(nG)

# 5 groups | Manhattan shows up most
```

```python
# Are private rooms the most popular in manhattan?

pR = air_bnb.groupby(['neighbourhood_group','room_type']).count().sort_values('neighbourhood_group', ascending=False)
pR

#No.. entire home/apt is. 
```

```python
# Which hosts are the busiest and based on their reviews?

busiest = air_bnb.groupby(['host_name','reviews_per_month']).count().sort_values('reviews_per_month', ascending=False)
busiest

#Row NYC is busiest based on average reviews per month
```

```python
#Which neighorhood group has the highest average price?
hAp = air_bnb.groupby(['neighbourhood_group']).mean()['price']
hAp

#Manhattan
```

```python
# Which neighbor hood group has the highest total price?
hTp = air_bnb.groupby(['neighbourhood_group']).sum()['price']
hTp
# Manhattan
```

```python
#Which top 5 hosts have the highest total price?
tFhTp = air_bnb.groupby(['host_name']).sum()['price'].sort_values(ascending=False)
tFhTp
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
busies = air_bnb[air_bnb['availability_365'] < 1]
busie = busies[busies['number_of_reviews'] > 100]
busie
```

```python
# What host has the highest total of prices and where are they located?

hTpw = air_bnb.groupby(['host_name','neighbourhood']).sum()['price'].sort_values(ascending=False)
hTpw
```

```python
# When did Danielle from Queens last receive a review?

hTp = air_bnb[air_bnb['host_name']=='Danielle']
hTc = hTp[hTp['neighbourhood']=='Brooklyn']
hTc
```
